```python
import folium
import numpy as np
import pandas as pd
```

# This is the map plotting of the European Airport Network, without edges

```python
# Create a map centered on Europe
map = folium.Map(location=[50, 15], zoom_start=4)
map_Berlin_2019 = folium.Map(location=[52.5200, 13.4050], zoom_start=12)
map_Berlin_2022 = folium.Map(location=[52.5200, 13.4050], zoom_start=12)
```

```python
# Load the data containing the locations' names, latitude, and longitude
locations_2019 = pd.read_csv("EU_routes_2019 (2).csv.gz")
locations_2022 = pd.read_csv("EU_routes_2022.csv.gz")
```

```python
locations_2022
```

```python
unique_values = np.unique(locations_2019["origin_country"])
print(unique_values)
```

```python
locations_2019
```

```python
# Add markers for each location to the map
for index, row in locations_2019.iterrows():
    folium.Marker([row['origin_latitude'], row['origin_longitude']], popup=row['origin_airport_name']).add_to(map)
```

```python
#Adds edges between airports for 2019 map
for i, edge in Berlin_2019.iterrows():
    folium.PolyLine([[edge['origin_latitude'], edge['origin_longitude']],
                     [edge['destination_latitude'], edge['destination_longitude']]],
                    #weight=edge['weight'],
                    color='black',
                    opacity=0.3).add_to(map_Berlin_2019)
```

```python
#Adds edges between airports for 2022 map
for i, edge in Berlin_2022.iterrows():
    folium.PolyLine([[edge['origin_latitude'], edge['origin_longitude']],
                     [edge['destination_latitude'], edge['destination_longitude']]],
                    #weight=edge['weight'],
                    color='black',
                    opacity=0.3).add_to(map_Berlin_2022)
```
